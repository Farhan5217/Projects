# Formulas used in the code below



1) Mean squared error formula

2) Partial derivative w.r.t x 

3) Partial derivative w.r.t y



```python
def Gradient_Descent(x,y):
    
    m_current=0
    b_current=0
    iterations=1000
    learning_rate=0.08
    n=len(x)
    
    for i in range(iterations):
        y_predicted= (m_current*x) +b_current
        m_derivative= - (2/n)*sum(x*(y-y_predicted))
        b_derivative= - (2/n)*sum(y-y_predicted)
        m_current= m_current-(learning_rate *m_derivative)
        b_current= b_current-(learning_rate *b_derivative)
        cost_function=(1/n)*sum((y-y_predicted)**2)
        print(f"m={m_current}, b={b_current}, Iteration Num={i}, Cost Function={cost_function},")

x=np.array([1,2,3,4,5])
y=np.array([5,7,9,11,13])

    
```

```python
Gradient_Descent(x,y)
```
